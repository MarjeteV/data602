```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python

# Question 1
sns.histplot(data=titanic, x='age', bins=20)
plt.title('Distribution of Ages in Titanic Dataset')
plt.show()
```

```python
#Question 2
missing_age_values = titanic['age'].isnull().sum()
print (missing_age_values) #checking how many were missing

titanic_age_dropna = titanic['age'].dropna()
sns.histplot(data=titanic_age_dropna, bins=20)
plt.title('Distribution of Ages in Titanic Dataset (no NAs)')

# Show the plot
plt.show()
```

```python
#3. After 'bins=... comma 'color='..')
sns.histplot(data=titanic, x='age', bins=20, color='purple')
```

```python
#4 scatter plots are use to compare two continous variables

# scatter plot to compare 'age' and 'fare'
sns.scatterplot(data=titanic, x='age', y='fare')

# title
plt.title('Comparison of Age and Fare')

# Show the plot
plt.show()

```

```python
#5 Correlation plot

numeric_titanic = titanic.select_dtypes(include=['float64', 'int64']) #gender caused an error before adding this code

# Compute the correlation matrix
corr_matrix = numeric_titanic.corr()

# Plot correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Plot of Numerical Variables in Titanic Dataset')

# Show the plot
plt.show()
```

```python
#6 change size - adding "plt.figure(figsize=(#, #)"" can be used on scatter plot ect as well.

# Plot the correlation heatmap from code above with
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Plot of Numerical Variables in Titanic Dataset')

# Show the plot
plt.show()
```
